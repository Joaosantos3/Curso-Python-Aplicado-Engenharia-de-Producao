### Breve resumo sobre o pyplot, figure e axes


O matplotlib.pyplot mantém um conceito de "figura atual" e "eixos atuais". Quando você usa funções como plt.plot(), plt.xlabel(), plt.title(), etc., essas funções aplicam-se à figura e aos eixos que estão atualmente ativos.

- **Axis**: ax é uma instância da classe Axes do matplotlib. Ele representa um único gráfico dentro de uma Figure. É dentro dos Axes que os dados são realmente plotados. Cada Axes tem seus próprios eixos (X e Y), título, rótulos, etc.

- **Figure**: representa a janela ou a área de desenho na qual os gráficos são desenhados. Você pode pensar na Figure como um contêiner que pode conter um ou mais gráficos (Axes). É comum usar 'fig' ao criar gráficos mais complexos ou ao trabalhar com múltiplos subplots.

Usamos instâncias Axis e Figure para ⭐personalizar⭐ gráficos e desenhos.

Por exemplo, para pintar o fundo do eixo de prata:

```python
ax = plt.axes() #uso eixo como objeto
ax.set_facecolor('silver') #acesso propriedade/atributo
plt.plot(array_1,array_2,'r--',label="data1") #plotando
```

### Boxplot: Breve Explicação

#### Um boxplot, ou gráfico de caixa, é uma representação gráfica que resume a distribuição de um conjunto de dados. Ele mostra a mediana, quartis e possíveis outliers de forma visual e compacta. Aqui estão os principais elementos de um boxplot:

- Mediana: A linha no meio da caixa representa a mediana (valor central) dos dados.

- Quartis: As bordas da caixa representam o primeiro quartil (Q1, 25%) e o terceiro quartil (Q3, 75%), dividindo os dados em quatro partes iguais.

- Intervalo Interquartil (IQR): A distância entre Q1 e Q3. Mede a dispersão dos dados.

- Whiskers: As linhas que se estendem dos quartis até os valores mínimos e máximos dentro de 1,5 vezes o IQR.

- Outliers: Pontos fora dos whiskers, indicando valores anômalos.

**Os boxplots são úteis para identificar a distribuição, dispersão e outliers de um conjunto de dados de forma rápida e visual.**

### Subplots no matplotlib

#### A função subplots no matplotlib é usada para criar uma figura contendo múltiplos gráficos (ou eixos) organizados em uma grade (grid). Isso é útil quando você deseja comparar diferentes conjuntos de dados ou visualizar várias métricas em um layout organizado.

#### A função plt.subplots() retorna uma tupla contendo uma figura (Figure) e uma matriz de objetos de eixos (Axes). A quantidade de eixos depende da configuração da grade especificada.

Os parâmetros:

- nrows: Número de linhas na grade de subplots.
- ncols: Número de colunas na grade de subplots.

```python
fig, axs = plt.subplots(nrows, ncols)
```

Faremos a plotagem dos três gráficos (BoxPlot,Histograma,Barchart) em uma única figura usando subplots.

Para acrescentá-los usaremos as próprias funções de plotagem do pandas (```boxplot```,```hist```,```bar```). No parâmetros ```ax=``` definimos qual ax receberá a plotagem. Exemplo, considerando o subplot trabalha em uma única direção (possui uma dimensão, isto é, possui várias linhas e 1 coluna ou 1 coluna e várias linhas):

```python
fig, axs = plt.subplots(2,1)
df.boxplot(by='Grupo',column='Cadencia',ax=axs1[0])
```
Para uma matriz de plotagens, quando número de linhas e colunas superam 1, temos:

```python
fig, axs = plt.subplots(2,2)
df.boxplot(by='Grupo',column='Cadencia',ax=axs1[0,0])
```

### ☝ Extra: Transformando gráficos soltos em um Grid, semelhante a Dashboard

#### Aqui, usaremos a biblioteca matplotib, não apenas pyplot (apelidamos de plt).

#### Criaremos uma Figura (```Figure```) Usaremos os objetos ```GridSpec``` e adicionando subplots em um GridSpec "pai" ou "mestre".

Vamos criar as grades e locais para os axes:

```python
figure=plt.figure(figsize=(10,8),constrained_layout=True)

'''Criamos um gridspec mestre de 2x2'''
gs_master=mlp.gridspec.GridSpec(2,2,figure)

'''Criamos um gridspec menor 1x1 que estará na posição linha (0) e ocupa todas as colunas disponíveis do gridspec (:)'''
#Boxplot
gs_1=mlp.gridspec.GridSpecFromSubplotSpec(1,1,subplot_spec=gs_master[0,:])
box_axes=figure.add_subplot(gs_1[0])

'''Criamos um gridspec menor 3x2 que estará na posição linha (1) e ocupa todas as colunas disponíveis do gridspec (:)'''
#Barchart
gs_2=mlp.gridspec.GridSpecFromSubplotSpec(3,2,subplot_spec=gs_master[1,:])
bar_axes=figure.add_subplot(gs_2[:,0])

'''Adicionamos subplots no grid à direita, coluna 1 do grid anterior (3x2) para os histogramas. Precisaremos de 3, pois o pandas retornas três histogramas'''

#Histogramas
hist1_axes=figure.add_subplot(gs_2[0,1])
hist2_axes=figure.add_subplot(gs_2[1,1])
hist3_axes=figure.add_subplot(gs_2[2,1])
```
Para adicionar as plotagens aos axes, faremos para cada axe e plotagem aquilo de já fizemos anteriormente nos subplots, usamos o parâmetros ```ax=``` e nomeamos o axes

````python
figure=plt.figure(figsize=(10,8),constrained_layout=True)

gs_master=mlp.gridspec.GridSpec(2,2,figure)

#Boxplot
gs_1=mlp.gridspec.GridSpecFromSubplotSpec(1,1,subplot_spec=gs_master[0,:])
box_axes=figure.add_subplot(gs_1[0])

#Barchart
gs_2=mlp.gridspec.GridSpecFromSubplotSpec(3,2,subplot_spec=gs_master[1,:])
bar_axes=figure.add_subplot(gs_2[:,0])

#Histogramas
hist1_axes=figure.add_subplot(gs_2[0,1])
hist2_axes=figure.add_subplot(gs_2[1,1])
hist3_axes=figure.add_subplot(gs_2[2,1])
````
